```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering


#load dataset
df=pd.read_csv("/content/Mall_Customers.csv")


#select numerical column for clustering
#Replace these column for clustering
x=df[['Annual Income (k$)','Spending Score (1-100)']]

#create Agglomerative clustering model
agg=AgglomerativeClustering(n_clusters=5)

#fit  model and assign labels
df['cluster'] =agg.fit_predict(x)


#plot custering
plt.scatter(x['Annual Income (k$)'],x['Spending Score (1-100)'],c=df['cluster'],cmap='rainbow')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Customer Segmentation')
plt.show()

df.to_csv("agglo_clustered_data.csv",index=False)
print("Agglomerative clustered dataset saved as 'agglo_clustered_data.csv'")
```
